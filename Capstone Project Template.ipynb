##### Data Enrichment
Additonal transformation step to parse the `I94_SAS_Labels_Description.SAS` file to derive corresponding port city and state for the I94 Immigration data port field.
This will allow us to join the `I94 Immigration` data with the other datasets through the port code and city fields.

```python
with open("./I94_SAS_Labels_Descriptions.SAS") as f:
    content = f.readlines()
splitted_content = [x.split('=') for x in content[303:962]]
port_codes = [x[0].strip('\t').strip().strip("'") for x in splitted_content]
port_locations = [x[1].strip().strip("'").strip() for x in splitted_content]
port_city = [x.split(",")[0] for x in port_locations]
port_states = [x.split(",")[-1] for x in port_locations]
df_port_locations = pd.DataFrame({"port_code" : port_codes, "port_city": port_city, "port_state": port_states})
df_port_locations.head()
df_port_locations.to_csv('I94_SAS_Labels_Descriptions.csv')
```
